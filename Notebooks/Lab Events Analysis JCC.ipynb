```python
import pandas as pd
```

```python
LABEVENTS = "../../Resources/LABEVENTS.csv"
D_LABITEMS = "../../Resources/D_LABITEMS.csv"
PATIENTS = "../../Resources/PATIENTS.csv"
```

```python
lab_df = pd.read_csv(LABEVENTS)

lab_df.head()
```

```python
items_df = pd.read_csv(D_LABITEMS)

items_df = items_df.drop(['ROW_ID'], axis=1)

items_df.head()
```

```python
joined_df = lab_df.merge(items_df, on='ITEMID')

joined_df.head()
```

```python
blood_df = joined_df[joined_df.FLUID == "Blood"]

blood_df
```

```python
tests = ['Chloride, Whole Blood','Glucose', 'Hematocrit, Calculated', 'Hemoglobin','Lactate', 'Potassium, Whole Blood', 
         'Sodium, Whole Blood', 'Anion Gap', 'Albumin','Bicarbonate', 'Bilirubin, Total', 'Creatinine', 'Chloride', 
        'Magnesium','Phosphate', 'Potassium', 'Sodium', 'Urea Nitrogen', 'Hematocrit',
         'Platelet Count', 'PTT', 'INR(PT)', 'PT', 'White Blood Cells']
```

```python
# Blood Gas labels
blood_gas = blood_df[blood_df.CATEGORY == 'Blood Gas']
blood_gas = blood_gas[['SUBJECT_ID','HADM_ID','VALUENUM','LABEL','VALUEUOM','CATEGORY']]
#blood_gas.to_csv("../../../blood_gas.csv")
#blood_gas_chloride = blood_gas[blood_gas.LABEL == 'Chloride, Whole Blood']
#blood_gas_glucose = blood_gas[blood_gas.LABEL == 'Glucose']
#blood_gas_hematocritcalc = blood_gas[blood_gas.LABEL == 'Hematocrit, Calculated']
#blood_gas_hemoglobin = blood_gas[blood_gas.LABEL == 'Hemoglobin']
#blood_gas_lactate = blood_gas[blood_gas.LABEL == 'Lactate']
#blood_gas_potassiumwb = blood_gas[blood_gas.LABEL == 'Potassium, Whole Blood']
#blood_gas_sodiumwb = blood_gas[blood_gas.LABEL == 'Sodium, Whole Blood']

blood_gas = blood_gas[blood_gas['LABEL'].isin(tests)]

blood_gas_labels = blood_gas.drop_duplicates(subset='LABEL')

blood_gas_labels
```
