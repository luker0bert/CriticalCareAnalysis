```python
import pandas as pd
```

```python
labevents = pd.read_csv('../Resources/labevents.csv')
```

```python
labitems = pd.read_csv('../Resources/d_labitems.csv')
```

```python
labevents.head()
```

```python
labitems = labitems.drop(['ROW_ID'], axis=1)
labitems.head()
```

```python
labs = labevents.merge(labitems, on = 'ITEMID')
labs.head()
```

```python
set(labs['CATEGORY'])
```

```python
labs['CATEGORY'] = labs['CATEGORY'].apply(lambda x: x.title())
```

```python
labs['TEST'] = labs['CATEGORY'] + "_" + labs['LABEL']
labs.head()
```

```python
labs2 = pd.pivot_table(labs, index=['SUBJECT_ID','HADM_ID'],values='VALUENUM',columns='TEST')
labs2.head()
```
