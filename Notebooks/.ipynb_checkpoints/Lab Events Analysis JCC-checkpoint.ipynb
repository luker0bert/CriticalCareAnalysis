```python
import pandas as pd
```

```python
LABEVENTS = "../../Resources/LABEVENTS.csv"
D_LABITEMS = "../../Resources/D_LABITEMS.csv"
PATIENTS = "../../Resources/PATIENTS.csv"
```

```python
lab_df = pd.read_csv(LABEVENTS)

lab_df.head()
```

```python
items_df = pd.read_csv(D_LABITEMS)

items_df = items_df.drop(['ROW_ID'], axis=1)

items_df.head()
```

```python
joined_df = lab_df.merge(items_df, on='ITEMID')

joined_df.head()
```

```python
joined_df['FLUID'].value_counts()
```

```python
patient_df = pd.read_csv(PATIENTS)

patient_df = patient_df.drop(['ROW_ID'], axis=1)

patient_df.head()
```

```python
final_df = joined_df.merge(patient_df, on='SUBJECT_ID')

final_df.head()
```

```python
blood_df = final_df[final_df.FLUID == "Blood"]

blood_df
```

```python
# Blood Gas labels
blood_gas = blood_df[blood_df.CATEGORY == 'Blood Gas']
blood_gas_chloride = blood_gas[blood_gas.LABEL == 'Chloride, Whole Blood']
blood_gas_glucose = blood_gas[blood_gas.LABEL == 'Glucose']
blood_gas_hematocritcalc = blood_gas[blood_gas.LABEL == 'Hematocrit, Calculated']
blood_gas_hemoglobin = blood_gas[blood_gas.LABEL == 'Hemoglobin']
blood_gas_lactate = blood_gas[blood_gas.LABEL == 'Lactate']
blood_gas_potassiumwb = blood_gas[blood_gas.LABEL == 'Potassium, Whole Blood']
blood_gas_sodiumwb = blood_gas[blood_gas.LABEL == 'Sodium, Whole Blood']
```

```python
# Chemistry labels
chemistry = blood_df[blood_df.CATEGORY == 'Chemistry']
chemistry_anion_gap = chemistry[chemistry.LABEL == 'Anion Gap']
chemistry_albumin = chemistry[chemistry.LABEL == 'Albumin']
chemistry_bicarbonate = chemistry[chemistry.LABEL == 'Bicarbonate']
chemistry_bilirubin_total = chemistry[chemistry.LABEL == 'Bilirubin, Total']
chemistry_creatinine = chemistry[chemistry.LABEL == 'Creatinine']
chemistry_chloride = chemistry[chemistry.LABEL == 'Chloride']
chemistry_glucose = chemistry[chemistry.LABEL == 'Glucose']
chemistry_magnesium = chemistry[chemistry.LABEL == 'Magnesium']
chemistry_phosphate = chemistry[chemistry.LABEL == 'Phosphate']
chemistry_potassium = chemistry[chemistry.LABEL == 'Potassium']
chemistry_sodium = chemistry[chemistry.LABEL == 'Sodium']
chemistry_urea_nitrogen = chemistry[chemistry.LABEL == 'Urea Nitrogen']
```

```python
# Hematology labels
hematology = blood_df[blood_df.CATEGORY == 'Hematology']
hematology_hematocrit = hematology[hematology.LABEL == 'Hematocrit']
hematology_hemoglobin = hematology[hematology.LABEL == 'Hemoglobin']
hematology_platelet_count = hematology[hematology.LABEL == 'Platelet Count']
hematology_ptt = hematology[hematology.LABEL == 'PTT']
hematology_inr(pt) = hematology[hematology.LABEL == 'INR(PT)']
hematology_pt = hematology[hematology.LABEL == 'PT']
hematology_white_blood_cells = hematology[hematology.LABEL == 'White Blood Cells']
hematology_wbc_count = hematology[hematology.LABEL == 'WBC Count']
```
