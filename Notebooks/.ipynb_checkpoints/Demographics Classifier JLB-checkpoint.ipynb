```python
ADMISSIONS = "../../Resources/ADMISSIONS.csv"
admissions_df = pd.read_csv(ADMISSIONS)
```

```python
admissions_df.head()
```

```python
# Extract day, month, year from DateStringYYYYMMDD
admissions_df['ADMITTIME'] = pd.to_datetime(admissions_df['ADMITTIME'])
admissions_df['DISCHTIME'] = pd.to_datetime(admissions_df['DISCHTIME'])

#format = '%Y%m%d %H%M%S'

admissions_df['admit_year'] = admissions_df['ADMITTIME'].dt.year
admissions_df['admit_month'] = admissions_df['ADMITTIME'].dt.month
admissions_df['admit_day'] = admissions_df['ADMITTIME'].dt.day
admissions_df['admit_time'] = admissions_df['ADMITTIME'].dt.time

admissions_df['disch_year'] = admissions_df['DISCHTIME'].dt.year
admissions_df['disch_month'] = admissions_df['DISCHTIME'].dt.month
admissions_df['disch_day'] = admissions_df['DISCHTIME'].dt.day
admissions_df['disch_time'] = admissions_df['DISCHTIME'].dt.time

admissions_df.head()

# NOTE!  from this we see that the years are actually garbage data
```

```python
admissions_df.info()
```

```python
admissions_df = admissions_df.drop(['DEATHTIME','LANGUAGE','EDREGTIME','EDOUTTIME','admit_year','disch_year'], axis=1)
```

```python
admissions_clean = admissions_df.dropna()
admissions_clean.info()
```

```python
# categorize time of day... for now dropping time

admissions_clean = admissions_clean.drop(['admit_time','disch_time','ADMITTIME','DISCHTIME'], axis=1)
```

```python
# get dummies
admissions_clean = pd.get_dummies(admissions_clean)
admissions_clean.head()
```

```python
# assignment and reshaping of data
data = admissions_clean.drop(["HOSPITAL_EXPIRE_FLAG"], axis=1)
target = admissions_clean["HOSPITAL_EXPIRE_FLAG"].values.reshape(-1, 1)
```

```python
# Create test and train sets
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
print(f"----train & test sizes-----")
print(f"train size: {len(X_train)}, keys: {len(X_train.keys())}")
print(f"test size: {len(X_test)}, keys: {len(X_test.keys())}")
```

```python
# Modeling with logistic regression
classifier = LogisticRegression()
```

```python
# Fit
classifier.fit(X_train, y_train.ravel())
```
