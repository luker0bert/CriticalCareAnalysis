```python
import pandas as pd
import datetime
```

```python
hematology = pd.read_csv('../Resources/hematology.csv')
chemistry = pd.read_csv('../Resources/chemistry.csv')
bloodgas = pd.read_csv('../Resources/blood_gas.csv')
```

```python
bloodgas.loc[bloodgas['LABEL']=='Hemoglobin','LABEL'] = 'Hemoglobin_Blood_Gas'
bloodgas.loc[bloodgas['LABEL']=='Glucose','LABEL'] = 'Glucose_Blood_Gas'
chemistry.loc[chemistry['LABEL']=='Glucose','LABEL'] = 'Glucose_Chemistry'
hematology.loc[hematology['LABEL']=='Hemoglobin','LABEL'] = 'Hemoglobin_Hematology'
```

```python
tests = ['Chloride, Whole Blood','Glucose_Chemistry', 'Hematocrit, Calculated', 'Hemoglobin_Blood_Gas','Lactate', 'Potassium, Whole Blood', 
         'Sodium, Whole Blood', 'Anion Gap', 'Albumin','Bicarbonate', 'Bilirubin, Total', 'Creatinine', 'Chloride', 
         'Glucose_Blood_Gas', 'Magnesium','Phosphate', 'Potassium', 'Sodium', 'Urea Nitrogen', 'Hematocrit', 'Hemoglobin_Hematology',
         'Platelet Count', 'PTT', 'INR(PT)', 'PT', 'White Blood Cells']
```

```python
df = pd.concat([hematology,chemistry,bloodgas], ignore_index=True)
```

```python
df.head()
```

```python
len(df)
```

```python
df = df[df['LABEL'].isin(tests)]
```

```python
len(df)
```

```python
df = df.sort_values('CHARTTIME', axis = 0)
```

```python
df.head(50)
```

```python
df.drop_duplicates(subset=['HADM_ID','LABEL'], inplace=True)
```

```python
df[df['HADM_ID']==145834.0]
```

```python
df['CHARTTIME'] = df['CHARTTIME'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S').date())
```

```python
df['FLAG'].fillna(value = 'normal', inplace = True)
```

```python
df.head()
```

```python
df_tidy = pd.pivot_table(df, index=['SUBJECT_ID','HADM_ID'],values='FLAG',columns='LABEL', aggfunc='first')
```

```python
df_tidy.shape
```

```python
df_tidy.head()
```

```python
df_tidy.fillna(value = 'no test', inplace = True)
```

```python
df_tidy.to_csv('../Resources/labs.csv')
```
