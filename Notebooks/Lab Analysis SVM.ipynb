```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
admissions = pd.read_csv("../../Resources/Admissions.csv")
labs = pd.read_csv("../../Resources/labs.csv")
patients = pd.read_csv("../../Resources/PATIENTS.csv")
```

```python
#admissions left merge labs on HADM ID
#admissions left merge patients on SUBJECT ID
df1= pd.merge(admissions,labs, on='HADM_ID')
df2= pd.merge(admissions,patients, on="SUBJECT_ID")
df3= pd.merge(df1,df2, on='HADM_ID')
df3.head()
```
